# 🎓 Student Performance Statistics

Exploratory data analysis on student exam scores, grouped by gender, test preparation, and more.

Dataset: [Student Performance on Exams](https://www.kaggle.com/datasets/spscientist/students-performance-in-exams)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Student_performance_data.csv") 
df.head()
```

```python
#Summary of all numeric columns
df.describe()
```

```python
#Average scores by gender
df.groupby('Gender')[['GPA']].mean()
```

```python
print(df.columns.tolist())
```

```python
df.groupby('Gender')['GPA'].mean().plot(kind='bar')
plt.title("Average GPA by Gender")
plt.ylabel("GPA")
plt.show()
```

```python
sns.boxplot(x='StudyTimeWeekly', y='GPA', data=df)
plt.title("GPA vs Weekly Study Time")
plt.xticks(rotation=50)
plt.show()
```

## 📚 GPA vs Tutoring

```python
sns.set_theme(style="whitegrid", palette="Blues")
sns.barplot(x='Tutoring', y='GPA', data=df)
plt.title("Average GPA by Tutoring Status")
plt.ylabel("Average GPA")
plt.xlabel("Tutoring")          
plt.show()
```

**Insight:** Student receiving tutoring average **~ 0.4 GPA** point higher than those who don't.

## 🏠 GPA vs Parental Support

```python
sns.barplot(x='ParentalSupport', y='GPA', data=df, palette="Greens")
plt.title("GPA by Parental Support")
plt.ylabel("Average GPA")
plt.xlabel("Parental Support")
plt.show()
```

**Insight:** Parental encouragement translates into a noticable GAP boost across the board.

## 🎒 GPA by Grade Class

```python
sns.barplot(x='GradeClass', y='GPA', data=df, palette="Purples")
plt.title("Average GPA by Grade Class")
plt.ylabel("GPA")
plt.xlabel("Grade Class")
plt.show()
```

**Insight:** GPA gradually declines in higher grade levels, hinting at increased curriculum difficulty.

## 🎓 GPA by Parental Education Level

```python
order = df.groupby('ParentalEducation')['GPA'].mean().sort_values().index
sns.barplot(y='ParentalEducation', x='GPA', data=df, order=order, palette="Oranges_r")
plt.title("GPA vs Parental Education Level")
plt.xlabel("Average GPA")
plt.ylabel("Parental Education")
plt.show()
```

```python
print(df['ParentalEducation'].unique())
print(df['GPA'].dtype, df['GPA'].head())
```

```python
education_map = {
    0: "Some High School",
    1: "High School",
    2: "Associate Degree",
    3: "Bachelor's Degree",
    4: "Master's Degree"
}

df['ParentalEducation'] = df['ParentalEducation'].map(education_map)
```

**Insight:** Students with more highly educated parents (Bachelor’s or above) show better academic performance overall.

## ⚽ GPA vs Extracurricular Participation

```python
sns.barplot(x='Extracurricular', y='GPA', data=df, palette="viridis")
plt.title("GPA by Extracurricular Involvement")
plt.xlabel("GPA")
plt.ylabel("Extracurricular (Yes/No)")
plt.show()
```

**Insight:** Students in extracurriculars post a modest but consistent GPA lift.

## 📌 Key Takeaways

- **Tutoring** and **parental support** yield the largest GPA gains.  
- **Parental education** level remains a strong predictor of academic success.  
- **Extracurricular involvement** adds a smaller but positive GPA bump.  
- GPA dips slightly in higher grade classes, suggesting tougher coursework.
